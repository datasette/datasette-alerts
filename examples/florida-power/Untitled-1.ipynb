```sqlite

create table t as 
with latest_commits as (
  select 
    commit_id,
    time,
    git_at(repo, commit_id, 'fplmaps.com/StormFeedRestoration.json') as data
  from git_log('scrape-florida-outages')
  where json_valid(data)
  --where path = 'fplmaps.com/StormFeedRestoration.json'
  limit 100
)
select distinct
  commit_id,
  time,
  value ->> '$.dateReported' as date_reported,
  value ->> '$.lastUpdated' as last_updated,
  value ->> '$.etr' as estimated_time_restoration,
  value ->> '$.customersAffected' as customers_affected,
  value ->> '$.Cause' as cause,
  value ->> '$.status' as status,
  value ->> '$.lat' as latitude,
  value ->> '$.lng' as longitude
from latest_commits
join json_each(data, '$.outages');
```

```sqlite
select count(*) from t;
select * from t limit 10;
```
